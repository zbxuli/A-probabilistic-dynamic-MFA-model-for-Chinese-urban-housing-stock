```python
import numpy as np
import pandas as pd
from scipy.stats import weibull_min

# Define the matrices and data
population = np.random.rand(20, 5000)
urbanization = np.random.rand(20, 5000)
floor_u = np.random.rand(20, 5000)
data = pd.DataFrame({'V1': np.arange(1, 120), 'V2': np.random.rand(119)})

```

```python
# Calculate urban population
population_u = population * 1000 * urbanization

# Calculate urban housing stock
stock_u = population_u * floor_u

# Extract relevant columns from data
area_complete = data['V2']
year_complete = data['V1']

# Add urban housing stock in 2012 to the stock_u matrix
stock_u_2012 = np.random.rand(20, 1)  # Replace with your actual data for 2012
stock_u = np.concatenate((stock_u_2012, stock_u), axis=1)

# Define matrices for simulation
complete_sim = np.zeros((207, 5000))  # 

output_sim = np.zeros((88, 5000))

```

```python
# Loop through each simulation
for i in range(5000):
    # Initialize variables for each simulation
    year_complete = data['V1'].copy()
    area_complete = data['V2'].copy()
    
    A = bootstrap_results[i][0]  # Use square brackets to access array elements
    B = bootstrap_results[i][1]  # Use square brackets to access array elements
    
    # Loop for years 2013-2015
    for j in range(2013, 2016):
        output = np.sum(area_complete * (
            weibull_min.cdf(j - year_complete, c=B, scale=A) -
            weibull_min.cdf(j - 1 - year_complete, c=B, scale=A)
        ))
        
        input_ = stock_u[j - 2011, i] - stock_u[j - 2012, i] + output
        area_complete = np.append(area_complete, input_)
        year_complete = np.append(year_complete, j)
        output_sim[j - 2012, i] = output
    
    # Loop for years 2016-2100
    for j in range(2016, 2101):
        Y = min((j - 2011) // 5 + 4, stock_u.shape[0] - 1)
        output = np.sum(area_complete * (
            weibull_min.cdf(j - year_complete, c=B, scale=A) -
            weibull_min.cdf(j - 1 - year_complete, c=B, scale=A)
        ))
        
        
        input_ = (stock_u[Y, i] - stock_u[Y - 1, i]) / 5 + output
        area_complete = np.append(area_complete, input_)
        year_complete = np.append(year_complete, j)
        output_sim[j - 2013, i] = output

    complete_sim[:, i] = area_complete

# Print or use the complete_sim and output_sim matrices as needed


```
